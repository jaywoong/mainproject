```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.read_excel('MER_삼성전자.xlsx')
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
df
```

```python
#배당비율 구하기위해서 전날주당배당금 추가
df['전날주당배당금'] = df['주당배당금'].shift(1)
# 액면분할 비율 구하기, if df['거래량'] == 0 이라면....
k = df[df['거래량'] == 0].iloc[0].loc['일자']
a = df[df['거래량'] == 0].iloc[0].loc['주당배당금']
b = df[df['거래량'] == 0].iloc[0].loc['전날주당배당금']
c = a/b
k
```

```python
#액면분할된 데이터 처리 
lst = ['종가','시가','대비','고가','저가','주당배당금','EPS','BPS']
lst2 = ['거래량','상장주식수']
for i in lst:
    df[i] = np.where(df['일자'] <= k , df[i]/c, df[i])
for j in lst2:
    df[j] = np.where(df['일자'] <= k , df[j]*c, df[j])
```

```python
#전날주당배당금 삭제
df = df.drop(columns=['전날주당배당금'], axis=1)
```

```python
df
```

```python
# atr데이터 처리
df_atr = df.iloc[:,[0,1,5,6]]
df_atr
```

```python
# atr데이터 처리
lst = []
atr = []
for i in range(0,len(df_atr)-1):
    a = df_atr.iloc[i,2] - df_atr.iloc[i,3]
    b = df_atr.iloc[i,2] - df_atr.iloc[i+1,1]
    c = df_atr.iloc[i,3] - df_atr.iloc[i+1,1]
    lst = [abs(a),abs(b),abs(c)]
    atr.append(max(lst))

# 2011/09/01 전날의 가격데이터가 없어, (금일의 고가 - 금일의 저가)를 적용
atr.append(df_atr.iloc[i,2] - df_atr.iloc[i,3])
```

```python
df['atr'] = atr
df
```

```python
df.to_excel('액면분할test2.xlsx')
```
